## 1 Introdução

A resolução da equação de equilíbrio dinâmico de estruturas considerando a não linearidade dos materiais é uma tarefa de elevado custo computacional. Devido a isso, diversas técnicas são propostas na literatura para diminuir o custo computacional gerado pela análise dinãmica no domínio do tempo (AYOUB et al, 2022). Segundo Chatterjee (2000), o *Proper Orthogonal Decomposition* (POD)  é um método poderoso e elegante de análise de dados que visa obter descrições aproximadas de processos de alta dimensão. Chamado por vezes de *Principal Component Analysis*, O POD tem sido utilizado para obter dados aproximados e de baixa dimensão de fluxos de fluidos turbulentos, vibrações estruturais, detecção de danos, entre outros.


Segundo Liang et al. (2002), o POD fornece uma base para a decomposição modal a partir de um conjunto de funções, com dados obtidos experimentalmente ou de simulações numéricas. Suas propriedades sugerem que este método fornece a base ideal para uso em várias aplicações, como as citadas anteriormente. As funções de base que ele produz são comumente chamadas de autofunções empíricas, base empírica, funções ortogonais empíricas, modos ortogonais próprios ou vetores de base.

A formulação consiste em desacoplar a equação de equilíbrio dinâmico em vários sistemas de um grau de liberdade, assim como na análise dinâmica modal. 

\begin{align}
x(z,t) &= \sum_{i=1}^{N}a_k(t)\phi(z)= \Phi a
\end{align}

onde
\begin{align*}
a_k&= \phi_k^Tx(z,t) \  \ (i=1,2,3,...,N),\\
a &= (a_1,a_2,a_3,...,a_N)^T, \\
\Phi &= [\phi_1,\phi_2,\phi_3,...,\phi_N]
\end{align*}


Pode ser utilizado o erro médio quadrático como medida de otimização do problema.

\begin{align}
E\{||x-x(l)||^2\} \ e \ \  E\{||x-\hat{x}(l)||^2\}
\end{align}

em que $x(l)$ é a expressão aproximada do vetor x usando os primeiros $l$ vetores bases das bases ortonormais indeterminadas e $\hat{x}(l)$ é a expressão aproximada de x usando a base $l$ em $R^N$.

O objetivo do POD é achar uma sequência de vetores bases que satisfazem o seguinte problema de extremos.

\begin{align*}
\mbox{min }  \epsilon^2(l) &= E\{||x-x(l)||^2\}\\
\mbox{suj. a } \phi_i^T\phi_j &= \delta_{ij} \ \ i,j = 1,2,...,N.
\end{align*}

Onde $x(l) = \sum_{k=1}^{l}a_k\phi_k (l\le m) $ e o valor esperado de x é tomado como zero.





## 2 *Principal Component Analysis* (PCA)

O PCA é um método estatístico que identifica a estrutura através de uma observação estocástica do evento visando obter uma descrição compacta da mesma. O PCA pode ser visto como um método de maximização da variância.

Supondo que $x\in R^N$ é um vetor aleatório e $a_1,a_2,...,a_N \in R$ são os componentes principais de $x$.Para calcular a primeira componente $a_1$, considera-se que a mesma será uma combinação linear de cada elemento de $x$.

\begin{align}
a_1 = \sum_{i=1}^{N}\alpha_{i1}x_i = \alpha_1^Tx
\end{align}

onde $\alpha = (\alpha_{11},\alpha_{21},...,\alpha_{N1})^T$ é um vetor constante.

A variância de $a_1$ é

\begin{align}
\sigma_{a_1}^2 = \alpha_1^TE\{(x-E\{x\})(x-E\{x\})^T\} \alpha_1
\end{align}

em que 

\begin{align}
\Sigma_x = E\{(x-E\{x\})(x-E\{x\})^T\} 
\end{align}

É a matriz de covariãncia do vetor $x$ e $E\{\cdot\}$ é o operador de valor esperado. Para que seja possíble maximizar a variância de $x$, é necessário normalizar o vetor $\alpha_1$ de forma que $\alpha_1^T \alpha_1 = 1$. Portanto, o problema de maximizaçao pode ser escrito como

\begin{align*}
\mbox{máx }  \sigma^2_{\alpha_1} &= \alpha^T_1 \Sigma_x \alpha_1 \\
\mbox{suj. a} \ \ \alpha_1^T \alpha_1 &= 1
\end{align*}

Introduzindo o multiplicador de Lagrange $\lambda_1$, tem-se

\begin{align}
L(\alpha_1,\lambda_1) = \alpha_1^T\Sigma_x\alpha_1 + \lambda_1(1-\alpha^T_1\alpha_1).
\end{align}

Diferenciando em relação a $\alpha_1$

\begin{align}
\frac{\partial L(\alpha_1,\lambda_1)}{\partial \alpha_1} = 2\left(\Sigma_x-\lambda_1I\right)\alpha_1.
\end{align}

Sendo o segundo termo da equação igual a zero, chega-se num problema de autovalores e autovetores.

\begin{align}
\Sigma_x\alpha_1 = \lambda_1\alpha_1.
\end{align}

Dessa forma, para a maximização do problema, $\lambda_1$ deve ser escolhido como o maior autovalor de $\Sigma_x$.

Definida a primeira componente, é possível realizar o mesmo procedimento para os demais modos ortogonais, onde é perceptível, Segundo Lyang et al. (2002), que os próximos modos principais também serão autovetores da matriz de covariância $\Sigma_x$.

De modo geral, é possível escrever o método PCA como o seguinte problema de maximização

\begin{align*}
\mbox{máx }  \sum_{i=1}^{l}\sigma^2_{\alpha_i} &= \sum_{i=1}^l\alpha^T_i \Sigma_x \alpha_i \\
\mbox{suj. a} \ \ \alpha_1^T \alpha_1 &= \delta_{ij}
\end{align*}

E o erro médio quadrado pode ser escrito como

\begin{align}
\epsilon^2(l) = \sum_{i = l+1}^N \sigma_{a_i}^2 = \sum_{i = l+1}^N \lambda_i
\end{align}

Segundo Ayoub et al. (2022), o número de modos necessários para simular adequadamente o sistema pode ser estabelicido quando $p \ge 99\%$, onde P pode ser definido como

\begin{align}
p=\frac{ \sum_{i}^l \lambda_i}{ \sum_{j}^N \lambda_j} 
\end{align}

De forma prática, os vetores bases ortonormais podem ser calculados a partir da resolução do problema de autovalores e autovetores da matriz de correlação ($x^Tx$), ordenados de acordo com seus respectivos autovetores ($\lambda_i$) em ordem decrescente.
















## 3 Aplicação num sistema com 5 graus de liberdade com não linearidade física

```python
#------------------------------------------------------------------------------
# 1 Import necessary modules
#------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from plot_config import *
plot_config()
import scipy.linalg as sc
```

```python
#------------------------------------------------------------------------------
# 2 Newmark-Newton-Raphson Analysis
#------------------------------------------------------------------------------
def NLNewmarkM(k,m,c,p,u0,v0,t,e,fel):

# 0.0 iniciando os parâmetros

    nt  = int(len(t))
    n   = int(len(p[:,0]))
    a   = np.zeros((n,nt))
    v   = np.zeros((n,nt))
    u   = np.zeros((n,nt))
    fs  = np.zeros((n,nt))
    kt  = np.zeros((nt,n,n))
    kht = np.zeros((nt,n,n))
    ph  = np.zeros((n,nt))
    R   = np.zeros((n,nt))
#1.0 Determinação dos parâmetros iniciais:
    b = 0.25
    y = 0.50

#1.1 passo temporal

    dt = t[1] - t[0]
    
#1.2 variáveis de estado    
    kt[0] = k
    fs[:,0] = np.dot(kt[0],u[:,0]) 
    
#1.3 cálculo da aceleração inicial:

    a[:,0] = np.linalg.inv(m)@(p[:,0] - c@v[:,0] - fs[:,0])
            
# 1.4 Determinação das consntantes:
            
    a1 = 1/(b*dt**2)*m + y/(b*dt)*c
            
    a2 = 1/(b*dt)*m + (y/b -1)*c
            
    a3 = (1/(2*b)-1)*m + dt*(y/(2*b)-1)*c
    
#2.0 início da iteração temporal
    for i in range(nt-1):
# 2.1 iniciação dos valores
            
            u[:,i+1]  = u[:,i]
            fs[:,i+1] = fs[:,i]
            kt[i+1]   = kt[i]

# 2.2 Cálculo da força equivalente 

            ph[:,i+1] = p[:,i+1] + a1@u[:,i] + a2@v[:,i] + a3@a[:,i]
            
# 3.0 início da iteração de Newton-Raphson
            j = 0
            
            while True:

                R[:,i+1] = ph[:,i+1] - fs[:,i+1] - a1@u[:,i+1]
                
                j+=1
                
                if j>5000:
                    
                    print('erro: não converge na iteração',i)
                    break
                    
                if np.linalg.norm(R[:,i+1]) <= e:
                    
                    break
                    print('Numero de iterações:',j)
                else:
                    
                    kht[i+1] = kt[i+1] + a1
                
                    du = np.linalg.inv(kht[i+1])@R[:,i+1]
                    
                    u[:,i+1] += du
                    
                    kt[i+1],fs[:,i+1],sd0 =fel(u[:,i+1])
                    
                    #kt[i+1] = kt[i]
                    #fs[:,i+1] = np.dot(kt[i+1],u[:,i+1])

# 4.0 Determinação da velocidade e da aceleração

            v[:,i+1] = y/(b*dt)*(u[:,i+1] - u[:,i]) + (1-y/b)*v[:,i] + dt*(
                       1-y/(2*b))*a[:,i]
            
            a[:,i+1] = 1/(b*dt**2)*(u[:,i+1] - u[:,i]) - 1/(b*dt)*v[:,i] - (
                       1/(2*b) - 1)*a[:,i]
            
    return u,v,a,kt,fs,R
```

```python
#------------------------------------------------------------------------------
# 3 Defining nonlinearity of restitutive forces
#------------------------------------------------------------------------------     
def felnl(U):
    n  = len(U)
    sd = np.zeros(n)
    fd = np.zeros(n)
    Fd = np.zeros(n)
    k  = np.zeros(n)
    
#1.0 Cálculo do Story drift
    sd[0] = U[0]
    for i in range(1,n):
        sd[i] = U[i] - U[i-1]
#2.0 Determinação do estado
    for i in range(n):
        
        if abs(sd[i])<=1.25:
            k[i]  = 100
            fd[i] = k[i]*sd[i]
        else:
            #print('estrutura plastificada')
            k[i]  = 5 
            fd[i] = np.sign(sd[i])*(125 + k[i]*(abs(sd[i])-1.25))
            
#3.0 Montagem da Matriz de rigidez 

    K = np.array([[k[0]+k[1],-1*k[1],0,0,0],
                 [-1*k[1],k[1]+k[2],-1*k[2],0,0],
                 [0,-1*k[2],k[2]+k[3],-1*k[3],0],
                 [0,0,-1*k[3],k[3]+k[4],-1*k[4]],
                 [0,0,0,-1*k[4],1*k[4]]])
    
#4.0 Montagem do vetor de forças 
    Fd[4] = fd[4]
    
    for i in range(4):
        Fd[i] = fd[i]-fd[i+1]
        
    return K,Fd,sd
```

```python
#------------------------------------------------------------------------------
# 4 Problem's data
#------------------------------------------------------------------------------  
#1.0 Valores iniciais do problema
e = 1e-3
t = np.arange(0,15,1e-2)
k = 100
u0 = v0 = np.zeros(5)
K = k* np.array([[2,-1,0,0,0],
                 [-1,2,-1,0,0],
                 [0,-1,2,-1,0],
                 [0,0,-1,2,-1],
                 [0,0,0,-1,1]])

M = 0.2591* np.diag(np.ones(5))

C= 0.01* np.array([[69.01,-19.81,-3.395,-1.37,-0.873],
                 [-19.81,65.7,-21.18,-4.268,-2.243],
                 [-3.395,-21.18,64.83,-22.05,-5.638],
                 [-1.37,-4.268,-22.05,63.46,-25.45],
                 [-0.873,-2.243,-5.638,-25.45,43.65]])

#2.0 Calculo do vetor de forças

ug          = np.zeros((1,len(t)))
#ug[0]       = -0.5*386.1*np.sin(2*np.pi*t) 
ug[0]       = 3000 * np.random.randn(len(t)) *np.exp(-0.1*t)
#ug[0,10:21] = np.zeros(11)
P           = -M@np.ones((5,1))
P           = np.dot(P.round(4),ug)

#2.1 Plotagem do vetor de forças

plt.figure(1,figsize =(5*1.6180,5))
plt.plot(t,P[0,:],'k')
plt.title("System's external force")
plt.xlabel('Time (s)')
plt.ylabel('F (kips)')
plt.grid(axis = 'y')
```

```python
#------------------------------------------------------------------------------
# 5 Solving by time-history integration
#------------------------------------------------------------------------------  
u,v,a,kt,fs,R = NLNewmarkM(K,M,C,P,u0,v0,t,e,felnl)

# Plotagem do deslocamento no último nó

plt.figure(2,figsize =(5*1.6180,5))
plt.plot(t,u[-1],'k')
plt.title("Last node response using full model")
plt.xlabel('Time (s)')
plt.ylabel('u (in)')
plt.grid(axis = 'y')
plt.plot()
```

```python
#------------------------------------------------------------------------------
# 6 Getting POD modes and plotting the energy percentage
#------------------------------------------------------------------------------  

lamb,POD = sc.eig((1/(len(t)-1))*u@u.T)
ilamb    = np.flip(lamb.argsort())
lamb     = np.real(lamb[ilamb])
POD      = POD[:,ilamb]
# Calcular os autovalores e autovetores
w21,Phi1 = sc.eig(K,M)

iw   = w21.argsort()
w21  = w21[iw]          ## Garantindo a ordem dos autovalores e autovetores
Phi1 = Phi1[:,iw]
wr   = np.real(w21)
wk   = np.sqrt(wr)
fk   = (wk/(2*np.pi))

f2 = plt.figure(3, figsize=(16,6))

n = len(u[:,0])
x  = np.arange(0, 12, 2)/12
for k in range(n):
    
    qk     = np.zeros(n+1)
    qk[1:] = Phi1[:,k]
    qk    /= np.max(np.abs(qk))   
    

    qk2     = np.zeros(n+1)
    qk2[1:] = POD[:,k]
    qk2    /= np.max(np.abs(qk2))   
    
    plt.subplot(1,n,k+1)
    plt.plot(qk, x*10,color ='k',label = 'Eigenmodes')
    plt.plot(qk2, x*10,'k:',label = 'POD Modes')
    
    plt.ylim( 0.0, 0.8*10);
    plt.xlim(-1.5,  1.5);  plt.ylabel('Mode {}'.format(str(k+1)));
   
    plt.grid(True)
plt.legend(loc = 'lower right')
plt.savefig('podmodes_with external forces.pdf')
y = np.zeros(n)
y[0] = lamb[0]/np.sum(lamb)

for i in range(1,n):
    
    y[i] = np.sum(lamb[0:i+1])/np.sum(lamb)
    
plt.figure(4,figsize = (6*1.6180,6))
plt.plot(np.arange(n)+1,y*100,'k-o')
plt.ylabel('Energy ratio (\%)')
plt.xlabel('Modo')
plt.ylim(np.min(y*100),100)
plt.grid(True)
plt.savefig('energiaacumulada.pdf')
```

```python
#------------------------------------------------------------------------------
# 7 Solving non-linear equation using POD modes
#------------------------------------------------------------------------------  
def NLNewmarkpod(k,m,c,p,u0,v0,t,e,fel,POD):

# 0.0 iniciando os parâmetros

    nt  = int(len(t))
    a   = np.zeros(nt)
    v   = np.zeros(nt)
    u   = np.zeros(nt)
    fs  = np.zeros(nt)
    kt  = np.zeros(nt)
    kht = np.zeros(nt)
    ph  = np.zeros(nt)
    R   = np.zeros(nt)
    
#1.0 Determinação dos parâmetros iniciais:
    b = 0.25
    y = 0.50

#1.1 passo temporal

    dt = t[1] - t[0]
    
#1.2 variáveis de estado    
    kt[0] = POD.T@k@POD
    fs[0] = POD.T@k@POD * u[0] 
    
#1.3 cálculo da aceleração inicial:

    a[0] = (POD.T@p[:,0] - POD.T@c@POD*v0 - fs[0])/(POD.T@m@POD)
            
# 1.4 Determinação das constantes:
            
    a1 = 1/(b*dt**2)*POD.T@m@POD + y/(b*dt)*POD.T@c@POD
            
    a2 = 1/(b*dt)*POD.T@m@POD + (y/b -1)*POD.T@c@POD
            
    a3 = (1/(2*b)-1)*POD.T@m@POD + dt*(y/(2*b)-1)*POD.T@c@POD
      
#2.0 início da iteração temporal
    for i in range(nt-1):
# 2.1 iniciação dos valores
            
            u[i+1]  = u[i]
            fs[i+1] = fs[i]
            kt[i+1] = kt[i]
            
# 2.2 Cálculo da força equivalente 
            ph[i+1] = POD.T@p[:,i+1] + a1*u[i] + a2*v[i] + a3*a[i]
            
# 3.0 início da iteração de Newton-Raphson
            
            k = 0
            while k<1:

                R[i+1] = ph[i+1] - fs[i+1] -a1*u[i+1]
                
                if abs(R[i+1]) <= e:
                    k = 1
                else:
                    
                    kht[i+1] = kt[i+1] + a1
                
                    du       = R[i+1]/kht[i+1]
            
                    u[i+1]  += du
            
                    kt[i+1],fs[i+1] = fel(POD*u[i+1],POD)


# 4.0 Determinação da velocidade e da aceleração

            v[i+1] = y/(b*dt)*(u[i+1] - u[i]) + (1-y/b)*v[i] + dt*(1-y/(2*b))*a[i]
            
            a[i+1] = 1/(b*dt**2)*(u[i+1] - u[i]) - 1/(b*dt)*v[i] - (1/(2*b) - 1)*a[i]
            
    return u,v,a,kt,fs,R
                    
#------------------------------------------------------------------------------
# 8 Defining nonlinearity of restitutive forces
#------------------------------------------------------------------------------     
def felnl2(U,POD):
    n  = len(U)
    sd = np.zeros(n)
    fd = np.zeros(n)
    Fd = np.zeros(n)
    k  = np.zeros(n)
    
#1.0 Cálculo do Story drift
    sd[0] = U[0]
    
    for i in range(1,n):
        sd[i] = U[i] - U[i-1]
#2.0 Determinação do estado    
    for i in range(n):
                   
        if abs(sd[i])<=1.250:
            k[i]  = 100
            fd[i] = k[i]*sd[i]
        else:
            
            k[i]  = 5 
            fd[i] = np.sign(sd[i])*(125 + k[i]*(abs(sd[i])-1.25))
            
#3.0 Montagem da Matriz de rigidez 

    K = np.array([[k[0]+k[1],-1*k[1],0,0,0],
                 [-1*k[1],k[1]+k[2],-1*k[2],0,0],
                 [0,-1*k[2],k[2]+k[3],-1*k[3],0],
                 [0,0,-1*k[3],k[3]+k[4],-1*k[4]],
                 [0,0,0,-1*k[4],1*k[4]]])
    
#4.0 Montagem do vetor de forças 
    Fd[4] = fd[4]
    
    for i in range(4):
        Fd[i] = fd[i]-fd[i+1]
        
    return POD.T@K@POD,POD.T@Fd      
```

```python
#------------------------------------------------------------------------------
# 9 Solving System using only the first POD mode
#------------------------------------------------------------------------------   
d,v,a,kt,fs,R = NLNewmarkpod(K,M,C,P,0,0,t,e,felnl2,POD[:,0])
PODuf         = POD[-1,0]*d

plt.figure(5,figsize = (6*1.6180,6))
plt.plot(t,PODuf,'k',label ='$1^{rst}$ mode Reduced model')
plt.plot(t,u[-1],'k--',label = 'Full model')
plt.xlabel('Time (s)')
plt.ylabel('u (in)')
plt.legend()
plt.grid(axis = 'y')
plt.savefig('comparacao.pdf')


d2,v,a,kt,fs,R = NLNewmarkpod(K,M,C,P,0,0,t,e,felnl2,POD[:,1])
d3,v,a,kt,fs,R = NLNewmarkpod(K,M,C,P,0,0,t,e,felnl2,POD[:,2])
d4,v,a,kt,fs,R = NLNewmarkpod(K,M,C,P,0,0,t,e,felnl2,POD[:,3])
d5,v,a,kt,fs,R = NLNewmarkpod(K,M,C,P,0,0,t,e,felnl2,POD[:,4])

PODuft         = POD[-1,0]*d + POD[-1,1]*d2 + POD[-1,2]*d3 + POD[-1,3]*d4 + POD[-1,4]*d5

plt.figure(6,figsize = (6*1.6180,6))
plt.plot(t,PODuft,'k',label ='5 Modes Reduced model')
plt.plot(t,u[-1],'k--',label = 'Full model')
plt.xlabel('Time (s)')
plt.ylabel('u (in)')
plt.legend()
plt.grid(axis = 'y')
```
